```python
import gmaps
import gmaps.datasets
from config import gmaps_api
import numpy as np
import pandas as pd
gmaps.configure(api_key=gmaps_api)
```

```python
breweries_csv = pd.read_csv('Code/brewery_data.csv')
breweries_df2 = pd.DataFrame(breweries_csv)
breweries_df2.head()
```

```python
# Need to get rid of NaN lat and lon in order to plot
breweries_val2 = breweries_df2.dropna(subset=['latitude', 'longitude'])
breweries_val2.head()
```

```python
breweries_val2 = breweries_val2[['latitude', 'longitude']]
breweries_layer = gmaps.symbol_layer(
    breweries_val2, fill_color='green', stroke_color = 'green', scale=2
)
fig = gmaps.figure()
fig.add_layer(breweries_layer)
fig
```
